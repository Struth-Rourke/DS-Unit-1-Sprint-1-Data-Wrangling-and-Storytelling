```python
df = pd.read_csv('LoanStats_2018Q4.csv', skiprows = 1, skipfooter = 2)

print(df.shape)
df.head()
```

```python
##### Term Objective #####

term_length = df['term'][0] # Creating a variable to see what the cell contents is like

term_length
```

```python
term_length.strip().strip(' months') # Stripping the unnecessary information out of the cell to just get the value
```

```python
term_length = int(term_length.strip().strip(' months')) # Saving the new stripped value over the old variable

term_length
```

```python
type(term_length) # Confirming that the new value is indeed an integer
```

```python
term_length_2 = df['term'][1] # New test variable to see whether or not the function I wrote next is correct

term_length_2
```

```python
# New function and calling that function for the new variable created above

def string_to_integer(cell_contents):
  return int(cell_contents.strip().strip(' months'))

string_to_integer(term_length_2)
```

```python
type(string_to_integer(term_length_2)) # Checking that the output is indeed an integer
```

```python
df['term'] = df['term'].apply(string_to_integer) # Using the x.apply function to replace all the old cell contents with the new integer content

df['term'].dtype # Confirming that the dtype for the 'Term' column is indeed a integer
```

```python
df['term'] # Proof of change; bottom right dtype = int64
```

```python
##### Loan Status is Great Objective #####

print(df['loan_status'][0]) # Checking to see what index contains what content 
print(df['loan_status'][1])
print(df['loan_status'][3])
```

```python
# New function to change the contents of each index to the new values ( 1 or 0)

def loan_status_change(cell_contents):
  if cell_contents == 'Fully Paid' or cell_contents == 'Current':
    return int(1)
  else:
    return int(0)

print(loan_status_change(df['loan_status'][0])) 
print(loan_status_change(df['loan_status'][1]))
print(loan_status_change(df['loan_status'][3]))
```

```python
df['loan_status_is_great'] = df['loan_status'].apply(loan_status_change) # Applying the function to the entirety of the Loan Status column

df['loan_status_is_great'].dtypes # Checking the dtype of the column to make sure it made the change throughout the column; was object, is now integer 
```

```python
df['loan_status_is_great'][:10] # Checking to make sure the change was made throughout the column
```

```python
df.head() # Check that Loan Status is Great is a new column inserted in the dataframe
```
